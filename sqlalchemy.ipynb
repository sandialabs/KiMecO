```python
from genericpath import isfile
from sqlalchemy import create_engine, Engine, URL
from sqlalchemy_utils import database_exists, create_database
import os

here = 'GAME'
print(here)
user = 'csoulie'
host = 'localhost'
# host = '146.246.246.69'
port = 5432
database = f'{here}/testdb'
url_object = URL.create(drivername="postgresql",
                 username=user,
                 host=host,
                 port=port,
                 database=database)
eng = create_engine(url_object)
print(eng.url.port)
print(eng.url)
```

```python

from sqlalchemy_utils import database_exists, create_database
if not database_exists(eng.url):
    print('creating db')
    create_database(eng.url)
    print('db created')
else:
    print('db exists')
```

```python
import socket
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
result = sock.connect_ex((host,port))
if result == 0:
   print("Port is open")
else:
   print("Port is not open")
sock.close()
```
